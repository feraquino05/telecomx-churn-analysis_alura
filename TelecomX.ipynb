```python
import pandas as pd
import requests

url = "https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/main/TelecomX_Data.json"

# Descargar el JSON correctamente
response = requests.get(url)
data = response.json()

# Ahora s normalizar correctamente
df = pd.json_normalize(data)

# Limpiar nombres de columnas
df.columns = df.columns.str.lower().str.replace('.', '_')

print(df.shape)
df.head()
```

```python
df.columns
```

```python
df['cuentas_diarias'] = df['account_monthlycharges'] / 30
df['cuentas_diarias'] = df['cuentas_diarias'].round(2)

df[['account_monthlycharges', 'cuentas_diarias']].head()
```

```python
df['cuentas_diarias'] = df['account_charges_monthly'] / 30
df['cuentas_diarias'] = df['cuentas_diarias'].round(2)

df[['account_charges_monthly', 'cuentas_diarias']].head()
```

```python
df.select_dtypes(include='object').nunique()
```

```python
# 1 Eliminar ID
df.drop(columns=['customerid'], inplace=True)

# 2 Convertir binarias
binary_cols = [
    'churn',
    'customer_partner',
    'customer_dependents',
    'phone_phoneservice',
    'account_paperlessbilling'
]

for col in binary_cols:
    df[col] = df[col].map({'Yes': 1, 'No': 0})

# 3 Convertir gnero
df['customer_gender'] = df['customer_gender'].map({'Male': 1, 'Female': 0})

# 4 Variables categricas mltiples
df = pd.get_dummies(df, drop_first=True)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Lista de variables categricas a explorar
categorical_cols = [
    'customer_gender',
    'account_contract',
    'account_paymentmethod',
    'phone_multiplelines',
    'internet_internetservice',
    'internet_onlinesecurity'
]

# Crear un grfico por cada columna
for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x=col, hue='churn', palette='coolwarm')
    plt.title(f'Distribucin de churn segn {col}')
    plt.xlabel(col)
    plt.ylabel('Cantidad de clientes')
    plt.legend(title='Churn', labels=['No', 'S'])
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
for col in categorical_cols:
    prop = df.groupby(col)['churn'].mean() * 100
    prop = prop.sort_values(ascending=False)

    plt.figure(figsize=(6,4))
    sns.barplot(x=prop.index, y=prop.values, palette='magma')
    plt.ylabel('% de churn')
    plt.xlabel(col)
    plt.title(f'Porcentaje de clientes que abandonaron segn {col}')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
